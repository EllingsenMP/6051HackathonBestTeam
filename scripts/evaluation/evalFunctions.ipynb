```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
dataEN = pd.read_csv('eduBot30_gemma_responses.csv')
dataES = pd.read_csv('eduBotSpanish30_gemma_responses.csv')
dataBUL = pd.read_excel('eduBot30_gemma_responses.xlsx')
```

```python
def text_cosine_similarity(responseOne, responseTwo):
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([responseOne, responseTwo])
    score = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
    return score[0][0]
```

```python
#Check Accuracy of responses to tutor
dataEN['accScore'] = dataEN.apply(lambda row: text_cosine_similarity(row['tutor_response'], row['gemma_response_1']), axis=1)
dataES['accScore'] = dataES.apply(lambda row: text_cosine_similarity(row['tutor_response'], row['gemma_response_1']), axis=1)
dataBUL['accScore'] = dataBUL.apply(lambda row: text_cosine_similarity(row['tutor_response'], row['gemma_response_1']), axis=1)

#Check adaptability of responses to tutor / first pass with subject
dataEN['adptScoreP1'] = dataEN.apply(lambda row: text_cosine_similarity(row['tutor_response'], row['gemma_response_2']), axis=1)
dataES['adptScoreP1'] = dataES.apply(lambda row: text_cosine_similarity(row['tutor_response'], row['gemma_response_2']), axis=1)
dataBUL['adptScoreP1'] = dataBUL.apply(lambda row: text_cosine_similarity(row['tutor_response'], row['gemma_response_2']), axis=1)

#Check adaptability of responses to tutor / final pass with subject + grade level
dataEN['adptScoreP2'] = dataEN.apply(lambda row: text_cosine_similarity(row['tutor_response'], row['gemma_response_3']), axis=1)
dataES['adptScoreP2'] = dataES.apply(lambda row: text_cosine_similarity(row['tutor_response'], row['gemma_response_3']), axis=1)
dataBUL['adptScoreP2'] = dataBUL.apply(lambda row: text_cosine_similarity(row['tutor_response'], row['gemma_response_3']), axis=1)

#Check stubbornness of responses across different passes / no context -> subject context
dataEN['stuScoreP1'] = dataEN.apply(lambda row: text_cosine_similarity(row['gemma_response_1'], row['gemma_response_2']), axis=1)
dataES['stuScoreP1'] = dataES.apply(lambda row: text_cosine_similarity(row['gemma_response_1'], row['gemma_response_2']), axis=1)
dataBUL['stuScoreP1'] = dataBUL.apply(lambda row: text_cosine_similarity(row['gemma_response_1'], row['gemma_response_2']), axis=1)

#Check stubbornness of responses across different passes / no context response -> subject + grade level context
dataEN['stuScoreP2'] = dataEN.apply(lambda row: text_cosine_similarity(row['gemma_response_1'], row['gemma_response_3']), axis=1)
dataES['stuScoreP2'] = dataES.apply(lambda row: text_cosine_similarity(row['gemma_response_1'], row['gemma_response_3']), axis=1)
dataBUL['stuScoreP2'] = dataBUL.apply(lambda row: text_cosine_similarity(row['gemma_response_1'], row['gemma_response_3']), axis=1)
```

```python
#ENGLISH
acc = dataEN['accScore'].mean()
adpt1 = dataEN['adptScoreP1'].mean()
adpt2 = dataEN['adptScoreP2'].mean()
stub1 = dataEN['stuScoreP1'].mean()
stub2 = dataEN['stuScoreP2'].mean()
```
